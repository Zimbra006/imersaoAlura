```python
dados_resultados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true')
dados_resultados.head()
```

```python
# Mostrar as top 10 aÃ§Ãµes dos moas
lista_moas = dados_resultados.columns.to_list()
lista_acoes = list()

for moa in lista_moas:
  lista_acoes.append(moa.split('_')[-1])

acoes = pd.Series(data = lista_acoes).value_counts().head(n = 10)

plt.figure(figsize = (12, 5))
ax = sns.barplot(x = acoes.index, y = acoes.values)
ax.set_title('Top 10 AÃ§Ãµes')
plt.show()
```

```python
# 0 significa nÃ£o ativado (False)
# ' significa ativado (True)
dados_resultados['5-alpha_reductase_inhibitor'].unique()
```

```python
# axis = 1 significa que vai retirar uma coluna
# moa = mecanismo de aÃ§Ã£o
contagem_moa = dados_resultados.drop('id', axis = 1).sum().sort_values(ascending = False)
contagem_moa
```

```python
#verificar se um determinado composto ativa mais de um mecanismo de aÃ§Ã£o
# axis = 1, ao invÃ©s de somar linha por linha pra cada coluna, ele soma coluna por coluna pra cada linha
dados_resultados.drop('id', axis = 1).sum(axis = 1)
```

```python
dados_resultados['n_moa'] = dados_resultados.drop('id', axis = 1).sum(axis = 1)
```

```python
dados_resultados['n_moa'] != 0
```

```python
dados_resultados['ativo_moa'] = (dados_resultados['n_moa'] != 0)
dados_resultados.head()
```

```python
dados_combinados = pd.merge(dados, dados_resultados[['id', 'n_moa', 'ativo_moa']], on = 'id')
dados_combinados.head()
```

```python
# Todos os testes de controle nÃ£o ativaram nenhum moa, o que Ã© esperado
dados_combinados.query('tratamento == "com_controle"')['ativo_moa'].value_counts()
```

```python
# Curiosamente, algumas drogas tambÃ©m nÃ£o despertaram nenhum moa
dados_combinados.query('tratamento == "com_droga"')['ativo_moa'].value_counts()
```

```python
plt.figure(figsize = (12, 8))
sns.boxplot(data = dados_combinados.query('composto in @top_compostos'), y = 'g-0', x = 'composto', hue = 'ativo_moa')
```

```python
dados_combinados['eh_controle'] = (dados_combinados['tratamento'] == 'com_controle')
dados_combinados.head()
```

```python
dados_combinados['24_horas'] = (dados_combinados['tempo'] == 24)
dados_combinados['48_horas'] = (dados_combinados['tempo'] == 48)
dados_combinados['72_horas'] = (dados_combinados['tempo'] == 72)
dados_combinados.head()
```

```python
# A partir dessa analisa, pode-se concluir que nenhum composto que nÃ£o Ã© controle ativa moas que nÃ£o foram ativados em outros testes
analise_ativo_moa = dados_combinados.query('not eh_controle').melt(id_vars = ['composto', 'dose', 'tempo'], value_vars = ['ativo_moa']).sort_values(by = ['composto', 'dose', 'tempo'], ignore_index = True)
valores_falsos = analise_ativo_moa.query('not value')['composto'].unique()
analise_ativo_moa.query('composto in @valores_falsos').query('value')
```

```python
analise_n_moas = dados_combinados.query('not eh_controle').melt(id_vars = ['composto', 'dose', 'tempo'], value_vars = ['n_moa']).sort_values(by = ['composto', 'dose', 'tempo'], ignore_index = True).query('value != 0')
#6 valores Ãºnicos para n_moa: 1, 2, 3, 4, 5, 7
analise_n_moas['value'].unique()
#Pelas anÃ¡lises feitas abaixo, vemos que nenhum composto ativa moas diferentes, independente do experimento
```

```python
valores_um = analise_n_moas.query('value == 1')['composto'].unique()
analise_n_moas.query('composto in @valores_um').query('value > 1')
```

```python
valores_dois = analise_n_moas.query('value == 2')['composto'].unique()
analise_n_moas.query('composto in @valores_dois').query('value != 2')
```

```python
valores_tres = analise_n_moas.query('value == 3')['composto'].unique()
analise_n_moas.query('composto in @valores_tres').query('value != 3')
```

```python
valores_quatro = analise_n_moas.query('value == 4')['composto'].unique()
analise_n_moas.query('composto in @valores_quatro').query('value != 4')
```

```python
valores_cinco = analise_n_moas.query('value == 5')['composto'].unique()
analise_n_moas.query('composto in @valores_cinco').query('value != 5')
```
