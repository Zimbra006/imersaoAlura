```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


# Separa os valores de teste e de treino do Machine Learning
# Os x's são os valores a partir dos quais a previsão
# Os y's são os valores nos quais as previsões deveriam chegar

# O stratify serve para que (x_treino / y_treino) = (x_teste / y_teste) = (x / y), ou seja, tudo seja o mais próximo da realidade quanto possível
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 376)

modelo_rlogistica = LogisticRegression(max_iter = 1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

```python
from sklearn.ensemble import RandomForestClassifier

x = dados_combinados.drop(['id', 'n_moa', 'ativo_moa', 'composto', 'tratamento'], axis= 1)
x = pd.get_dummies(x, columns = ['dose', 'tempo'])
y = dados_combinados['ativo_moa']

# Curiosamente, não importa o quanto eu mude test_size, o modelo acerta algo por volta de 68% todas as vezes
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.5, stratify = y, random_state = 376)

modelo_randomforest = RandomForestClassifier()
modelo_randomforest.fit(x_treino, y_treino)
modelo_randomforest.score(x_teste, y_teste)

# Após testar outras configurações, eu tentei retirar a coluna de tratamento da base de dados
# Ao fazer isso, a taxa de acertos chegou em 99%
```

```python
modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

```python
# rodar outros testes
# teste 01: ver se consegue prever se não ativa nenhum moa

x = dados_combinados.drop(['id', 'n_moa', 'ativo_moa', 'composto', 'tratamento'], axis= 1)
x = pd.get_dummies(x, columns = ['dose', 'tempo'])
y = dados_combinados['0_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.5, stratify = y, random_state = 376)

modelo_randomforest = RandomForestClassifier()
modelo_randomforest.fit(x_treino, y_treino)
modelo_randomforest.score(x_teste, y_teste)

# Conclusões: mesmo tirando a coluna tratamento, ele acerta em 99% das vezes
# Isso significa que é mais fácil retirar da base de dados os que ele acha que não ativaria nenhum moa
# do que tentar prever se o composto ativaria algum moa (isso muda ao retirar a coluna tratamento do teste acima)
```

```python
modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

```python
# teste 02: ver se consegue prever se ativa somente 1 moa

x = dados_combinados.drop(['id', 'n_moa', 'ativo_moa', 'composto', 'tratamento'], axis= 1)
x = pd.get_dummies(x, columns = ['dose', 'tempo'])
y = dados_combinados['1_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.5, stratify = y, random_state = 376)

modelo_randomforest = RandomForestClassifier()
modelo_randomforest.fit(x_treino, y_treino)
modelo_randomforest.score(x_teste, y_teste)

# Conclusões: nesse ele também acerta 99% das vezes
# Aparentemente, tentar deduzir se ele ativa qualquer moa é mais complicado
# do que tentar deduzir se ele ativa só um ou só dois, etc
```
