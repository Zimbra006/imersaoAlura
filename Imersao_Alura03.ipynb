```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
dados.head()
```

```python
#Retorna os top 5 compostos mais usados
#(value_counts() ordena eles dos mais pros menos usados)
#(index[n:m] mostra os valores nos espaços n até m, não incluído, no caso, os 5 primeiros)
top_compostos = dados['composto'].value_counts().index[0:5]
top_compostos
```

```python
plot_data = dados.query('composto in @top_compostos')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# devido a uns probleminhas com o collab,
# o .set() aqui embaixo precisa ser chamado pelo menos uma vez
sns.set()
sns.set_theme(font = 'sans-serif', font_scale = 1.5)
plt.figure(figsize=(10, 6))
# ax é o nome padrão para fazer gráficos
ax = sns.countplot(x = 'composto', data = plot_data, order = top_compostos)
ax.set_title('Top 5 Compostos')
plt.show()
```

```python
corr = dados.loc[:, 'g-0':'g-50'].corr()
```

```python
import numpy as np

sns.set_theme(font_scale = 1, style = 'white')
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
corr_celular = dados.loc[:, 'c-0':'c-50'].corr()
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_celular, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_celular, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
cell_gene_corr = dados.loc[:, 'g-0':'g-50'].merge(dados.loc[:, 'c-0':'c-50'], on = dados['id']).corr()
cell_gene_corr = cell_gene_corr.loc['g-0':'g-50', 'c-0':'c-50']
cell_gene_corr
```

```python
# Esconde os dados que são avaliados como verdadeiros pela 'mask'
mask = np.triu(np.ones_like(cell_gene_corr, dtype=bool))

# Retorna f, a figura com 11 x 9 polegadas de tamanho, e os eixos ax
f, ax = plt.subplots(figsize=(11, 9))

# Cria uma paleta de cores com os valores 230 e 20 sendo as âncoras negativas e positivas
# E retorna ela como um matplotlib.colors.ColorMap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Desenha o mapa de calor, escondendo a partir de 'mask' e usando q paleta de 'cmap'
# Centralizando os valores a partir de 0, deixando eles em caixas quadradas
sns.heatmap(cell_gene_corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

# Existem genes que, quando tem suas expressões muito altas, são inversos à sobrevivência celular
# Da mesma forma que existem genes que cooperam com a sobrevivência
# Tem até uns que não fazem diferença
```
